```python
import numpy as np
import matplotlib.pyplot as plt
from persim import PersistenceImager

pdgms = [np.array([[0.2, 0.95], [0.3, 0.4], [0.6,1.0]])]

# Create figure and subplots
fig, axs = plt.subplots(1, 5, figsize=(13, 4.5))

# Titles and labels
titles = [
    "Original Diagram",
    "Birth-Persistence\nCoordinates",
    "Persistence Image\n($\\sigma=0.01$)",
    "Persistence Image\n($\\sigma=0.03$)",
    "Persistence Image\n($\\sigma=0.07$)"
]
labels = ["(a)", "(b)", "(c)", "(d)", "(e)"]

# Plot content and set titles/labels
for i in range(5):
    # axs[i].set_title(titles[i])
    # Plotting logic
    if i == 0:
        pimgr.plot_diagram(pdgms[0], skew=False, ax=axs[i])
    elif i == 1:
        pimgr.plot_diagram(pdgms[0], skew=True, ax=axs[i])
    else:
        sigma = [0.01, 0.03, 0.07][i - 2]
        pimgr = PersistenceImager(birth_range=(0.0, 1.0), pixel_size=0.05)
        pimgr.kernel_params = {'sigma': sigma}
        pimgs = pimgr.transform(pdgms, skew=True)
        pimgr.plot_image(pimgs[0], ax=axs[i])
    
    # Add subplot label below the plot
    axs[i].text(0.5, -0.30, labels[i] + " " + titles[i], transform=axs[i].transAxes,
                ha='center', va='top', fontsize=12)

plt.tight_layout()
plt.show()
```
